```python
def get_access_token():

    # URLS
    AUTH_URL    = 'https://accounts.spotify.com/authorize'
    TOKEN_URL   = 'https://accounts.spotify.com/api/token'
    CLIENT_ID   = '17d35b6f7c6140969fbea4f258d4b72a'
    CLIENT_SECRET = '2e2d4cbba9b5468886b08d74c2767e5a'

    # Ver Autorization Code Flow: 'https://developer.spotify.com/documentation/general/guides/authorization/code-flow/'
    auth_params = {
        'client_id': CLIENT_ID,
        'response_type': 'code',
        'redirect_uri': 'http://localhost:8888/callback',
        'scope': 'user-read-private',
    }
        
    auth_code = requests.get(AUTH_URL, auth_params)
    auth_header = base64.urlsafe_b64encode(bytes(CLIENT_ID + ':' + CLIENT_SECRET, 'utf-8'))
    
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded',
        'Authorization': 'Basic %s' % auth_header.decode('ascii')
    }

    payload = {
        'grant_type': 'client_credentials',
        'code': auth_code,
        'redirect_uri': 'https://open.spotify.com/collection/playlists',
    }

    # request to /token endpoint. aquí obtenermos el token de acceso 
    access_token_request = requests.post(url=TOKEN_URL, data=payload, headers=headers)
    access_token_response_data = access_token_request.json()

    # extraemos y almacenamos únicamente el token del json
    access_token = access_token_response_data['access_token']

    return  access_token
```

```python
def search_artist(atoken, BASE_URL, name):
    pass
    search_headers = {
        'Content-Type': 'application/json',
        'Authorization': 'Bearer ' + atoken
    }
    search_pamans = {
        'q': name, 
        'type':'artist',
        'limit':50,
        'offset': 0,
        'include_external': 'FALSE'
        
    }

    new_search = requests.get( BASE_URL + 'search?', headers=search_headers, params=search_pamans).json()
    return new_search
```

```python
import os
def write_to_csv(asearch):
    
    csv_path = r'C:\Users\nelso\Desktop\ironhacktareas\Final_Project\Spotify_v2.2\Spotify_v2\temp_csv_files\_list_artist.csv'
    csv_data = pd.DataFrame()
    csv_data = pd.json_normalize(asearch['artists']['items'],max_level=1)
    csv_data = csv_data.reset_index(drop=True)
    
    if len(csv_data)==0:
        csv_data = csv_data.reset_index(drop=True)
    else:
        csv_data = csv_data.drop(columns=["images","uri","external_urls.spotify","followers.href"])
        csv_data = csv_data.reset_index(drop=True)
    #print(csv_data)            # imprime el dataframe en pantalla
    csv_data.to_csv(csv_path,index=False,mode='a')
    
    del asearch, csv_data
    
    return 0
```

```python
def clean_csv_duplicated():
    csv_path = os.getcwd() + '/temp_csv_files/'
    #csv_path = os.path.dirname(__file__)
    with open(r'C:\Users\nelso\Desktop\ironhacktareas\Final_Project\Spotify_v2.2\Spotify_v2\temp_csv_files\_list_artist.csv', 'r') as infile, open(r'C:\Users\nelso\Desktop\ironhacktareas\Final_Project\Spotify_v2.2\Spotify_v2\temp_csv_files\Final_list_artist.csv','w') as outfile:
        seen = set()
        for line in infile:
            if line in seen: continue
            
            seen.add(line)
            outfile.write(line)
        
        infile.close()
        outfile.close()
                
    del csv_path, infile, line, seen
    return 0
```

```python
def remove_csv_columns():
    #csv_path = os.path.dirname(__file__) + '/temp_csv_files/'
    with open(r'C:\Users\nelso\Desktop\ironhacktareas\Final_Project\Spotify_v2.2\Spotify_v2\temp_csv_files\Final_list_artist.csv', 'r') as infile, open(r'C:\Users\nelso\Desktop\ironhacktareas\Final_Project\Spotify_v2.2\Spotify_v2\temp_csv_files\Final_list_artist_v2.csv','w') as outfile:
        
        count = 0
        for line in infile:
            if line[0:3] == '[],':
                count = count + 1
            else:         
                outfile.write(line)
        infile.close()
        outfile.close()
                
    del  infile, line
    return 0
```

```python
def remove_csv_files():
    #csv_path = os.getcwd() + '/temp_csv_files/'
    os.remove(r'C:\Users\nelso\Desktop\ironhacktareas\Final_Project\Spotify_v2.2\Spotify_v2\temp_csv_files\_list_artist.csv')
            
    return 0
```

```python
def main():
    BASE_URL    = 'https://api.spotify.com/v1/'
    atoken = get_access_token()

    artist_name = get_artist_list()

    for a_name in artist_name:
        time.sleep(0.1)
        asearch = search_artist(atoken, BASE_URL, a_name)
        returned_value = write_to_csv(asearch)
        del asearch, returned_value
    returned_value = clean_csv_duplicated()
    time.sleep(0.01)
    returned_value = remove_csv_columns()
    returned_value = remove_csv_files()
    del returned_value
    return 0     
```

```python
main()
```
